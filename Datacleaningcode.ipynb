```python
#using the median to replace the missing values
median = df['total_confirmed'].median()
df['total_confirmed'].fillna(median, inplace=True)


#using the median to replace the missing values
median = df['total_deceased'].median()
df['total_deceased'].fillna(median, inplace=True)

#using the median to replace the missing values
median = df['total_recovered'].median()
df['total_recovered'].fillna(median, inplace=True)

#using the median to replace the missing values
median = df['total_tested'].median()
df['total_tested'].fillna(median, inplace=True)

#using the median to replace the missing values
median = df['population'].median()
df['population'].fillna(median, inplace=True)

```
